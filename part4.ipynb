# Probability Density Function (PDF) Of The Eigenvalue Distribution

```python
def eigenvalue_pdf(graph):
    adj_matrix = nx.adjacency_matrix(graph).toarray()
    return np.linalg.eigvals(adj_matrix)
```

# ER_LINE_GRAPH PDF Of Eigenvalue Distribution

```python
def draw_er_line_eigenvalue_pdf(graph):
    er_eigenvalue_pdf = eigenvalue_pdf(graph)

    plt.hist(er_eigenvalue_pdf, bins=50, density=True, alpha=0.5)
    plt.xlabel('Eigenvalues')
    plt.ylabel('Probability Density')
    plt.title('ER_LINE_GRAPH Eigenvalue Distribution PDF')
    plt.show()
```

```python
draw_er_line_eigenvalue_pdf(er_line_graph)
```

# ER_GRAPH PDF Of Eigenvalue Distribution

```python
def draw_er_eigenvalue_pdf(graph):
    er_eigenvalue_pdf = eigenvalue_pdf(graph)

    plt.hist(er_eigenvalue_pdf, bins=50, density=True, alpha=0.5)
    plt.xlabel('Eigenvalues')
    plt.ylabel('Probability Density')
    plt.title('ER GRAPH Eigenvalue Distribution PDF')
    plt.show()
```

```python
draw_er_eigenvalue_pdf(er_graph)
```

# SF_LINE_GRAPH PDF Of Eigenvalue Distribution

```python
def draw_sf_line_eigenvalue_pdf(graph):
    sf_eigenvalue_pdf = eigenvalue_pdf(graph)
    
    hist, bins = np.histogram(sf_eigenvalue_pdf, bins=100, density=True)
    
    # Compute the PDF
    pdf = hist / np.sum(hist)
    
    # Plot the PDF
    plt.bar(bins[:-1], pdf, width=np.diff(bins), align='edge')
    plt.xlabel('Eigenvalues')
    plt.ylabel('Probability Density')
    plt.title('SF Line Graph Eigenvalue Distribution PDF')
    plt.show()
```

```python
draw_sf_line_eigenvalue_pdf(sf_line_graph)
```

```python
num_of_simulation = 100

random_range = 201 


er_graphs = []
for _ in range(0,num_of_simulation):
        n = random.randint(1,random_range)
        pc = np.log(n) / n
        p = 2 * pc   
        erdosh_graph = nx.gnp_random_graph(n, p)

        er_graphs.append({
          "n": n,
          "p": p,
          "graph": erdosh_graph
        })
```

```python
er_table = []
for er in er_graphs:
        n = er['n']
        p = er['p']
        erdosh_graph = er['graph']

        er_table.append([n, p])
cols = ['n', 'p']
df = pd.DataFrame(er_table, columns=cols, index=range(1, 101))  
df.head(30)    
```
